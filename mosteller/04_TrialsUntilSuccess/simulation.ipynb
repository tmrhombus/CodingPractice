```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_iterations = 100000
```

```python
def roll_die():
    return random.randint(1,6)
```

```python
def rolls_until_6():
    n_rolls = 1
    while True:
        a = roll_die()
        if a==6:
            return n_rolls
        else:
            n_rolls+=1
```

```python
list_rolls = [ rolls_until_6() for i in range(n_iterations)  ]
avg_n_rolls = np.mean(list_rolls)
print(f"On average it takes {avg_n_rolls} rolls")
```

```python
figw = 10
figh = 5
nbins = 30

c_base = "#012155"
c_hist = "black"
fs = 20
lw = 3

title = "Histogram of Nr. Rolls until Six"
ytitle = "Counts"
xtitle = "Nr. Rolls"

fig, ax1 = plt.subplots(1, 1, figsize = (figw, figh))

ax1.set_title(title, fontsize=fs+5, loc="left", color=c_base)
ax1.set_ylabel(ytitle, fontsize=fs, color=c_base)
ax1.set_xlabel(xtitle, fontsize=fs, color=c_base)
ax1.tick_params( axis="both", which="major", labelsize=fs-5, colors=c_base )

ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.spines["bottom"].set_color(c_base)
ax1.spines["left"].set_color(c_base)

ax1.hist( list_rolls, bins=nbins, histtype="step", linewidth=lw, color=c_hist )

ax1.axvline( avg_n_rolls, lw=lw, color=c_hist, linestyle=":" )
```
